```python
# Preprocessing the data:
#
# 1. Remove the 'Production (tonnes)' string from columns' names as it
#    occupies significant space and that information is reduntant
#    throughout the columns
#
# 2. Replace the column name 'Entity' with 'Country' - It is more intuitive
#
# 3. The rest of the table is well organized and ready to be connected with SQL
```

```python
# 1.

import pandas as pd
import pymysql

df = pd.read_csv('../01_data_files/world_food_production_raw_data.csv')

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 10)
```

```python
df.columns.values
```

```python
remove_string = 'Production(tonnes)'

for i in range(len(df.columns)):
    
    df.columns.values[i] = df.columns.values[i].replace(' ', '')
    df.columns.values[i] = df.columns.values[i].replace(remove_string, '')
    
```

```python
df.columns.values
```

```python
# 2.
new_column_names = ['Country', 'Year', 'Maize', 'Rice', 'Yams', 'Wheat', 'Tomatoes',
       'Tea', 'Sweetpotatoes', 'Sunflowerseed', 'Sugarcane', 'Soybeans',
       'Rye', 'Potatoes', 'Oranges', 'Peas,dry', 'Palmoil', 'Grapes',
       'Coffee,green', 'Cocoabeans', 'Meat,chicken', 'Bananas',
       'Avocados', 'Apples']
df.columns = new_column_names
```

```python
df
```

```python
# Cote d'Ivoire as an ' as it would create an error in SQL syntax
# ' needs to be removed
for i in range(df.shape[0]):
    if ("'") in df[df.columns[0]][i]: 
        df[df.columns[0]][i] = df[df.columns[0]][i].replace("'", "")
```

```python
# Loop to create multipe query lines to be inserted in SQL subsequently

a = 0
b = 2000
k = 2000
lista = []

while a < df.shape[0]:    
    insert_query = 'INSERT INTO world_food_production VALUES '
    
    if a > 10000:
        new_df = df.iloc[b:,:]
        
    else:
        new_df = df.iloc[a:b,:]      

        for i in range(new_df.shape[0]):
            insert_query += '(' 

            for j in range(new_df.shape[1]):
                insert_query += "'" + str(new_df[new_df.columns[j]][i+a]) + "'" + ', '

            insert_query = insert_query[:-2] + '), '
        
    insert_query = insert_query[:-2] + ';'

    lista.append(insert_query)
    
    a += k
    b += k
```

```python
# 3.

conn = pymysql.connect(database= 'world_food_production', user = 'root', password = '365pass')
```

```python
cursor = conn.cursor()
```

```python
for i in range(len(lista)):
    cursor.execute(lista[i])
    conn.commit()
```
